# Generate README Example Images

This notebook generates all example images for the Vizard README.md.

**Instructions:**
1. Run all cells in order
2. Images will be saved to `../images/`
3. Check the output directory to verify all images were created

**Note:** Make sure you're in the test directory when running this notebook so relative paths work correctly.

```python
import polars as pl
import altair as alt
from pathlib import Path
```

```python
%cc VZ RESET
```

## Image 0: Most Simple Bar Chart

```python
%%cc
VZ
DATA data/sales.csv
PLOT bar
X product Y revenue
FILENAME images/most_basic_bar_chart_basic.png
```

```python
df = pl.read_csv('data/sales.csv')

chart = alt.Chart(df).mark_bar(color='steelblue').encode(
    x=alt.X('product:N', title='Product'),
    y=alt.Y('revenue:Q', title='Revenue')
).properties(width=600, height=400)

chart.save('images/most_basic_bar_chart_basic.png', scale_factor=2.0)

chart
```

## Image 1: Bar Chart - Product Revenue

```python
%%cc
RESET
DATA data/sales.csv
PLOT bar
X product Y revenue
COLOR category
TITLE Product Revenue by Category
WIDTH 600
HEIGHT 400
OUTPUT save
FILENAME images/bar_chart_basic.png
```

```python
df = pl.read_csv('data/sales.csv')

chart = alt.Chart(df).mark_bar().encode(
    x=alt.X('product:N', title='Product'),
    y=alt.Y('revenue:Q', title='Revenue'),
    color=alt.Color('category:N', title='Category')
).properties(
    title='Product Revenue by Category',
    width=600,
    height=400
)

chart.save('images/bar_chart_basic.png', scale_factor=2.0)
chart
```

## Image 2: Scatter Plot - Gene Expression

```python
%%cc
RESET
DATA data/genes.csv
PLOT scatter
X expression Y pvalue
COLOR significant
TITLE Gene Expression vs P-value
WIDTH 600
HEIGHT 400
OUTPUT save
FILENAME images/scatter_plot_basic.png
```

```python
df = pl.read_csv('data/genes.csv')

chart = alt.Chart(df).mark_point(size=60).encode(
    x=alt.X('expression:Q', title='Expression'),
    y=alt.Y('pvalue:Q', title='P-value'),
    color=alt.Color('significant:N', title='Significant')
).properties(
    title='Gene Expression vs P-value',
    width=600,
    height=400
)

chart.save('images/scatter_plot_basic.png', scale_factor=2.0)
chart
```

## Image 3: Line Chart - Temperature Time Series

```python
%%cc
DATA data/timeseries.csv
PLOT line
X date Y temperature
COLOR location
TITLE Temperature Over Time
OUTPUT save
DIR images
FILENAME line_chart_timeseries.png
```

```python
df = pl.read_csv('data/timeseries.csv')

chart = alt.Chart(df).mark_line().encode(
    x=alt.X('date:T', title='Date'),
    y=alt.Y('temperature:Q', title='Temperature'),
    color=alt.Color('location:N', title='Location')
).properties(
    title='Temperature Over Time',
    width=600,
    height=400
)

chart.save('images/line_chart_timeseries.png', scale_factor=2.0)
chart
```

## Image 4: Grouped Bar Chart - Gene Expression by Condition

```python
%%cc
DATA data/expression.csv
PLOT bar
X gene_name Y expression_level
COLOR condition
GROUP_TYPE grouped
TITLE Gene Expression by Condition
OUTPUT save
FILENAME images/bar_chart_grouped.png
```

```python
df = pl.read_csv('data/expression.csv')

chart = alt.Chart(df).mark_bar().encode(
    x=alt.X('gene_name:N', title='Gene Name'),
    xOffset=alt.XOffset('condition:N'),
    y=alt.Y('expression_level:Q', title='Expression Level'),
    color=alt.Color('condition:N', title='Condition')
).properties(
    title='Gene Expression by Condition',
    width=600,
    height=300
)

chart.save('images/bar_chart_grouped.png', scale_factor=2.0)
chart
```

## Image 5: Faceted Scatter Plot - Small Multiples

```python
%%cc
RESET
DATA data/data.csv
PLOT scatter
X value1 Y value2
COLOR condition
ROW replicate
TITLE Experimental Results by Replicate
WIDTH 500
HEIGHT 300
OUTPUT save
FILENAME images/scatter_faceted.png
```

%%cc
RESET
DATA data/data.csv
PLOT scatter
X value1 Y value2
COLOR condition
ROW replicate
TITLE Experimental Results by Replicate
WIDTH 500
HEIGHT 300
OUTPUT save
FILENAME docs/images/scatter_faceted.png

```python
df = (pl.read_csv('data/diff_expression.csv')
    .with_columns([
        pl.when((pl.col('log2fc') > 1.5) & (pl.col('neg_log10_pvalue') > 1.3))
          .then(pl.lit('upregulated'))
          .when((pl.col('log2fc') < -1.5) & (pl.col('neg_log10_pvalue') > 1.3))
          .then(pl.lit('downregulated'))
          .otherwise(pl.lit('not significant'))
          .alias('regulation')
    ]))

points = alt.Chart(df).mark_point(size=40).encode(
    x=alt.X('log2fc:Q', title='Log2 Fold Change'),
    y=alt.Y('neg_log10_pvalue:Q', title='-Log10(P-value)'),
    color=alt.Color('regulation:N',
                    scale=alt.Scale(domain=['upregulated', 'downregulated', 'not significant'],
                                    range=['red', 'blue', 'lightgray']),
                    legend=alt.Legend(title='Regulation'))
)

hline = alt.Chart(pl.DataFrame({'y': [1.3]})).mark_rule(strokeDash=[5,5], color='black').encode(y='y:Q')
vline1 = alt.Chart(pl.DataFrame({'x': [1.5]})).mark_rule(strokeDash=[5,5], color='black').encode(x='x:Q')
vline2 = alt.Chart(pl.DataFrame({'x': [-1.5]})).mark_rule(strokeDash=[5,5], color='black').encode(x='x:Q')

chart = (points + hline + vline1 + vline2).properties(
    title='Volcano Plot: Differential Gene Expression',
    width=500,
    height=300
)

chart.save('images/volcano_plot.png', scale_factor=2.0)
chart
```

## Image 7: Heatmap - Gene Expression Matrix

```python
%%cc
DATA data/expression_matrix.csv
PLOT heatmap
X sample Y gene
COLOR expression
Use viridis color scheme
TITLE Gene Expression Heatmap
OUTPUT save
FILENAME images/heatmap.png
```

```python
df = pl.read_csv('data/expression_matrix.csv')

chart = alt.Chart(df).mark_rect().encode(
    x=alt.X('sample:N', title='Sample'),
    y=alt.Y('gene:N', title='Gene'),
    color=alt.Color('expression:Q',
                    scale=alt.Scale(scheme='viridis'),
                    title='Expression')
).properties(
    title='Gene Expression Heatmap',
    width=500,
    height=300
)

chart.save('images/heatmap.png', scale_factor=2.0)
chart
```

## Image 8: Box Plot - Measurement Distributions

```python
%%cc
DATA data/measurements.csv
PLOT box
X group Y value
TITLE Measurement Distributions by Group
WIDTH 600
FILENAME images/box_plot.png
```

```python
df = pl.read_csv('data/measurements.csv')

chart = alt.Chart(df).mark_boxplot().encode(
    x=alt.X('group:N', title='Group'),
    y=alt.Y('value:Q', title='Value')
).properties(
    title='Measurement Distributions by Group',
    width=600,
    height=300
)

chart.save('images/box_plot.png', scale_factor=2.0)
chart
```

## Verify Images Were Created

```python
# Check which images were created
from pathlib import Path

output_dir = Path('images')
expected_images = [
    "bar_chart_basic.png",
    "scatter_plot_basic.png",
    "line_chart_timeseries.png",
    "bar_chart_grouped.png",
    "scatter_faceted.png",
    "volcano_plot.png",
    "heatmap.png",
    "box_plot.png"
]

print("Image Generation Status:")
print("=" * 50)
created_count = 0
for img in expected_images:
    img_path = output_dir / img
    if img_path.exists():
        size_kb = img_path.stat().st_size / 1024
        print(f"âœ“ {img:30s} ({size_kb:.1f} KB)")
        created_count += 1
    else:
        print(f"âœ— {img:30s} (MISSING)")

print("="  * 50)
print(f"Total: {created_count}/{len(expected_images)} images created")
print(f"Output directory: {output_dir.absolute()}")
```
