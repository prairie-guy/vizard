```python
%%cc create 12 samle data to use to test vizard. There are 4 sets of 3x replicates. Two of the sets are gDNA and two are dfDNA
Each set has two subsets, one is UMBS and the other EM-seq. Place this in a polars df. Each set of replicates should be 
similar, but with variance. 
```

```python
import polars as pl
import numpy as np

# Set seed for reproducibility
np.random.seed(42)

# Create the experimental design
samples = []
sample_id = 1

# Define base values for each group
base_values = {
    ('gDNA', 'UMBS'): 85.0,
    ('gDNA', 'EM-seq'): 92.0,
    ('cfDNA', 'UMBS'): 45.0,
    ('cfDNA', 'EM-seq'): 78.0
}

# Generate 3 replicates for each of the 4 combinations
for dna_type in ['gDNA', 'cfDNA']:
    for method in ['UMBS', 'EM-seq']:
        base_value = base_values[(dna_type, method)]
        
        # Create 3 replicates with variance (CV ~5-8%)
        for replicate in range(1, 4):
            value = base_value + np.random.normal(0, base_value * 0.06)
            
            samples.append({
                'sample_id': f'S{sample_id:02d}',
                'dna_type': dna_type,
                'method': method,
                'replicate': replicate,
                'measurement': round(value, 2)
            })
            sample_id += 1

# Create polars DataFrame
test_df = pl.DataFrame(samples)

test_df
```

```python
%cc adjust the sample names to very compactly encode the group and subgroups
```

```python
# Create compact sample names encoding group and subgroups
# Format: [DNA type initial][Method initial][replicate number]
# g = gDNA, c = cfDNA
# U = UMBS, E = EM-seq
# 1-3 = replicate number

test_df = test_df.with_columns(
    pl.concat_str([
        pl.when(pl.col('dna_type') == 'gDNA').then(pl.lit('g'))
          .otherwise(pl.lit('c')),
        pl.when(pl.col('method') == 'UMBS').then(pl.lit('U'))
          .otherwise(pl.lit('E')),
        pl.col('replicate').cast(pl.Utf8)
    ]).alias('sample_id')
)

test_df
```
